```julia
using Plots
using CSV
using DataFrames
using FreqTables
using Vega
```

```julia
csv_reader = CSV.File("winemag-data_first150k.csv")
df = DataFrame(csv_reader)
```

```julia
df_freq = freqtable(df.points)
bar_df = DataFrame(a=Array(df_freq), b=names(df_freq)[1])

plot(bar_df.b, bar_df.a, xrotation=90, label = "produced")
scatter!(bar_df.b, bar_df.a, label = "produced point")
```
